2.Which country has the highest number of billionares

```python
newdataset['country'].unique()#check the countries involved
```

```python
len(newdataset['country'].unique())#Know the total number of unique countries
```

```python
newdataset.groupby('country').size()#returns count of each unique country on the list 
```

```python
country_counts=newdataset.groupby('country').size()#saving it to a new variable
```

```python
country_counts=country_counts.to_frame()#convert data into dataframe
```

```python
country_counts=country_counts.rename(columns={0:'count'})#change the name of the count column from zero
```

```python
country_counts=country_counts.sort_values(by='count',ascending= False)#sort values in descending form
```

```python
country_counts
```

```python
fig,ax1=plt.subplots()#function that returns a tuple containing a figure and axes object(s)
ax1.bar(country_counts.index[:10],country_counts['count'][:10])
[s.set_visible(False) for s in ax1.spines.values()]
ax1.set_title('Countries per count of Billionare')
ax1.set_ylabel('Number of billionares per country')
ax1.set_xticklabels(labels=country_counts.index,rotation=90)
plt.grid(axis='y',linestyle='--',linewidth=0.5)
plt.show()
```

The 10 richest in Africa

```python
newdatsetafrica=newdataset.set_index(keys='country')
```

```python
newdatsetafrica=newdatsetafrica.loc[country_in_list]
```

```python
newdatsetafrica=newdatsetafrica.sort_values(by='NetWorth($ Billion)',ascending=False) 
```

```python
import seaborn as sns
```

```python
fig,ax5=plt.subplots(figsize=(7,5))
sns.barplot(newdatsetafrica['Name'][:10],newdatsetafrica['NetWorth($ Billion)'][:10],hue=newdatsetafrica.index[:10],dodge=False)
[s.set_visible(False) for s in ax5.spines.values()]
ax5.set_title('Top 10 richest in africa')
ax5.set_xlabel('Country')
ax5.set_xticklabels(newdatsetafrica['Name'][:10],rotation=90)
ax5.legend(fontsize=7,loc='upper right')
plt.grid(axis='y',linestyle='--',linewidth=0.3)
plt.show()
```

Distribution of Age
